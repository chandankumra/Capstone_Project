The population of each state

```python
driver.get('https://statisticstimes.com/demographics/india/indian-states-population.php')
```

```python
states=[]
st=driver.find_elements(By.XPATH,'//td[@class="name"]')
for i in st[0:36]:
    s=i.text
    states.append(s)
```

# Phase 4

Unsupervised ML (Clustering)



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012.csv')
df.head()
```

```python
df.shape
```

```python

dist=df['DISTRICT']
dist=pd.DataFrame({'DISTRICT':dist})
```

```python
dist
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
```

```python
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=oe.fit_transform(df[i].values.reshape(-1,1))
```

```python
df.head()
```

```python
data=df
```

```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=3,init='k-means++',random_state=25)
y_kmeans=kmean.fit_predict(data)
```

```python
np.unique(y_kmeans)
```

```python
y_kmeans
```

```python
test=kmean.predict(np.asarray([[220,6507]]))
test[0]
```

```python
data[y_kmeans==1]
```

```python
data.rename(columns={'TOTAL IPC CRIMES':'CRIMES'},inplace=True)
```

```python
import pickle
filename='final_stage.pickle'
pickle.dump(kmean,open(filename,'wb'))
```

```python
plt.figure(figsize=(10,12))
plt.scatter(data[y_kmeans==0]['DISTRICT'],data[y_kmeans==0]['CRIMES'],s=150,c='green',label='Peaceful Area')
plt.scatter(data[y_kmeans==1]['DISTRICT'],data[y_kmeans==1]['CRIMES'],s=150,c='red',label='Sensitive Area')
plt.scatter(data[y_kmeans==2]['DISTRICT'],data[y_kmeans==2]['CRIMES'],s=150,c='blue',label='Moderate Area')
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s=150,c='yellow',label='centroid')
plt.title('clusters of crimes')
plt.xlabel('district')
plt.ylabel('crimes')
plt.legend()
plt.show()
```
